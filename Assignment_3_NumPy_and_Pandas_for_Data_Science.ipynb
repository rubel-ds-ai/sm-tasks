```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
df_practice = df.copy()
print("Original missing values:", df_practice.isnull().sum())

# Step 2: Create missing data
np.random.seed(42)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50, replace=False)
missing_writing_idx = np.random.choice(df_practice.index, 30, replace=False)

# Continue the task...
# Your code here:

# 2. Randomly set 50 math scores to np.nan
df_practice.loc[missing_math_idx, 'math score'] = np.nan
print("\n\nAfter inserting NaN: \n")
print(df_practice[df_practice['math score'].isnull()].head(5))
# 3. Randomly set 30 writing scores to np.nan
df_practice.loc[missing_writing_idx, 'writing score'] = np.nan
print(df_practice[df_practice['writing score'].isnull()].tail(5))
```

```python
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

missing_values_before_fillna = df_practice.isnull().sum()
print(missing_values_before_fillna)

import warnings

# Suppress the specific FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

math_median = df_practice['math score'].dropna().median()
df_practice['math score'].fillna(math_median, inplace=True)

writing_median = df_practice['writing score'].dropna().median()
df_practice['writing score'].fillna(writing_median, inplace=True)

missing_values_after_fillna = df_practice.isnull().sum()
print(missing_values_after_fillna)
```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

# Your code here:
```

```python
# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
#    - Count of students
# Average scores by gender
avg_scores_by_gender = df_practice.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Count of students by gender
count_by_gender = df_practice['gender'].value_counts()

print("Average Scores by Gender:\n", avg_scores_by_gender)
print("\nStudent Count by Gender:\n", count_by_gender)
```

```python
# avg_scores_by_gender
```

```python
# Mean scores by test preparation course
avg_scores_by_prep = df_practice.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
avg_scores_by_prep
```

```python

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
#    - Show which subject benefits most from test prep

# Calculate improvement due to test preparation
prep_effect = avg_scores_by_prep.loc['completed'] - avg_scores_by_prep.loc['none']
most_improved_subject = prep_effect.idxmax()
prep_effect
```

```python
most_improved_subject
```

```python

print("Average Scores by Test Preparation Course:\n", avg_scores_by_prep)
print("\nScore Improvement from Test Prep:\n", prep_effect)
print(f"\nSubject that benefits most from test prep: {most_improved_subject}")
```

```python
# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance

# Average math score by parental education
math_by_education = df_practice.groupby('parental level of education')['math score'].mean()

# Rank education levels by math performance (descending)
ranked_math_by_education = math_by_education.sort_values(ascending=False)

print("Average Math Score by Parental Level of Education:\n", math_by_education)
print("\nRanked Education Levels by Math Score:\n", ranked_math_by_education)
```

```python
# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category

avg_score_by_lunch = df_practice.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
# avg_score_by_lunch
count_by_lunch = df_practice['lunch'].value_counts()

print(f"Average score by lunch type: {avg_score_by_lunch}\n")
print(f"Student Count by lunch type: {count_by_lunch}\n")
```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
#    - Show the difference between male and female average

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance
```

```python
# Average math score by gender
avg_math_by_gender = df_practice.groupby('gender')['math score'].mean()

# Difference between female and male average
math_diff = avg_math_by_gender['female'] - avg_math_by_gender['male']

print("Average Math Score by Gender:")
print(avg_math_by_gender)
print(f"\n Difference (Female - Male): {math_diff:.2f}")

# Who performs better?
better_math_gender = 'Female' if math_diff > 0 else 'Male'
print(f"\n {better_math_gender} students perform better in math.")
```

```python
# Calculate average scores for each subject
subject_averages = df_practice[['math score', 'reading score', 'writing score']].mean()

# Sort from highest to lowest
ranked_subjects = subject_averages.sort_values(ascending=False)

print("Average Scores by Subject:")
print(subject_averages)
print("\n Subjects Ranked by Performance:")
print(ranked_subjects)

# Best performing subject
top_subject = ranked_subjects.idxmax()
print(f"\n Students perform best in: {top_subject}")
```

```python
# Add total score column
df_practice['total score'] = df_practice['math score'] + df_practice['reading score'] + df_practice['writing score']

# Average total score by parental education
avg_total_by_education = df_practice.groupby('parental level of education')['total score'].mean()

# Rank education levels by total score
ranked_education = avg_total_by_education.sort_values(ascending=False)

print("Average Total Score by Parental Education Level:")
print(avg_total_by_education)
print("\n Education Levels Ranked by Student Performance:")
print(ranked_education)

# Best-performing education level
top_education = ranked_education.idxmax()
print(f"\n Highest student performance is seen when parents have: {top_education}")
```
