```python
# Upload the Gym data in google drive

import pandas as pd

gym_dataset_url = "https://github.com/rubel-ds-ai/public-datasets/raw/refs/heads/main/sm-datasets/gym_membership.csv"

gym_dataset = pd.read_csv(gym_dataset_url)


# Display last 5 rows
gym_dataset.tail()
```

```python
# Task 1: Basic Matplotlib (3 plots)
# a. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()
import matplotlib.pyplot as plt
import seaborn as sns
gym_dataset['abonoment_type'].value_counts()
```

```python
value_count = gym_dataset['abonoment_type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(value_count.index, value_count.values)
plt.title("Gym Membership Types")
plt.xlabel("Membership Types")
plt.ylabel("Membership Counts")

plt.show()
```

```python
# Task 1: Basic Matplotlib (3 plots)
# b. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])
# gym_dataset['avg_time_in_gym'].plot(
#     kind='hist',
#     bins= 20,
#     figsize=(6,4),
#     color='purple',
#     edgecolor = 'black'
# )
gym_dataset['avg_time_in_gym'].plot.hist(       # .plot(kind='hist') == .plot.hist()
    bins= 20,
    figsize=(6,4),
    color='purple',
    edgecolor = 'black',
)
plt.title("Distribution of Time Spent in Gym")
plt.xlabel("Average Time in Gym (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Task 1: Basic Matplotlib (3 plots)
# c. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

gym_dataset.plot(
    kind = 'scatter',
    x = 'Age',
    y = 'visit_per_week',
    figsize = (6,4),
    color = 'darkorange',
    alpha = 0.75,
    title = 'Age Vs Visits per week',
    xlabel = 'Age',
    ylabel = 'Visit per week'
)
plt.show()
```

```python
# using seaborn
plt.figure(figsize=(6,4))
sns.scatterplot(data=gym_dataset, x="Age", y="visit_per_week", color="purple", alpha=0.6, edgecolor="k")
plt.title("Age vs. Visits per Week")
plt.show()
```

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

plt.figure(figsize=(6,4))
sns.boxplot(data=gym_dataset, x='gender', y='Age', palette='Set2')
plt.title('Box Plot of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python

# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')
plt.figure(figsize=(6,4))
sns.countplot(data=gym_dataset, x='days_per_week', palette='pastel')
plt.title('Count of Gym Visits per Week')
plt.xlabel('Days per Week')
plt.ylabel('Count')
plt.xticks(rotation=45)  # rotate x labels for readability
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=gym_dataset, y='days_per_week', palette='pastel')
plt.title('Count of Gym Visits per Week')
plt.xlabel('Count')
plt.ylabel('Days per Week')
plt.show()
```

```python
# Count occurrences
counts = gym_dataset['days_per_week'].value_counts()

# Keep only top 5, group the rest as 'Other'
top_days = counts.nlargest(5).index
gym_dataset['days_grouped'] = gym_dataset['days_per_week'].apply(lambda x: x if x in top_days else 'Other')

# Plot
plt.figure(figsize=(8,4))
sns.countplot(data=gym_dataset, x='days_grouped', palette='pastel', order=top_days.tolist()+['Other'])
plt.show()
```
